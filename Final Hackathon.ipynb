```python
# Importing the useful librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Circle
```

```python
#Reading train data [1.Customer data 2.Email data 3.Customer Retention Potential data 4.Transaction data]
data=[pd.read_excel('Customerdata_Train.xlsx'),pd.read_excel('Emaildata_Train.xlsx'),pd.read_excel('Train.xlsx'),pd.read_excel('transactionsdata_Train.xlsx')]
```

```python
# Observing the shape of the train data files
for i in data:
    print(i.shape)
```

```python
# Merging the train data files
data1=pd.merge(data[0],data[1],on='CustomerID')
data2=pd.merge(data[2],data1,on='CustomerID')
data=pd.merge(data[3],data2,on='CustomerID')
```
